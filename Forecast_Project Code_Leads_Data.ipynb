## 1. Importing libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
import datetime
```

## 2. Loading data

```python
qualified_leads_raw=pd.read_csv('olist_marketing_qualified_leads_dataset.csv')
closed_deals_raw=pd.read_csv('olist_closed_deals_dataset.csv')
```

```python
closed_deals_raw.head()
```

## 3. Loading copy and renaming columns

```python
qualified_leads=qualified_leads_raw.copy()
closed_deals=closed_deals_raw.copy()
```

```python
## checking data volume (rown, columns)

qualified_leads.shape
```

```python
closed_deals.shape  ## you can see that the conversion rate is a little higher than 10%
```

```python
### Renaming columns

qualified_leads=qualified_leads.rename(columns={'mql_id':'Marketing Lead ID','origin':'origem'})
closed_deals=closed_deals.rename(columns={'mql_id':'Marketing Lead ID', 'sr_id':'sales representative','sdr_id':'sales development'})
```

```python
closed_deals.head()
```

## 4. Create BigTable

```python
#TableName=activar biblioteca.merge
# on=nomedacoluna que vai conectar
# how = tipo de joint, nesse caso eu quero as 8000 linhas dos leads e os que nao converteam vao ficar sem os dados

tabelao=pd.merge(qualified_leads,closed_deals,on='Marketing Lead ID', how='left')
```

```python
tabelao.shape
```

```python
tabelao.head()
```

```python
tabelao.to_excel('tabelao_funil_marketing.xlsx')
```

```python
## colocar as datas em mes (first contact_date e won_date)

tabelao['month_year_first_content'] = pd.to_datetime(tabelao['first_contact_date']).dt.to_period('M')
tabelao['month_year_won_date'] = pd.to_datetime(tabelao['won_date']).dt.to_period('M')
```

## 5. Convert BigTable in a Excel File

```python
tabelao.to_excel('tabelao_funil_marketingvf.xlsx')
```
